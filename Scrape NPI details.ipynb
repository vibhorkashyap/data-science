```python
import csv
from bs4 import BeautifulSoup
import requests
import json
```

```python
mbr_list = []
with open(r'C:\Users\vkashya5\Desktop\mbr.csv','r')as f:
    data = csv.reader(f)
    for row in data:
        mbr_list.append(row[0])
```

```python
the_list = []
failed_npi = []
for npi in mbr_list:
    try:
        url = 'http://www.hipaaspace.com/medical_billing/coding/national_provider_identifier/codes/npi_'+npi+'.json'
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
        js = soup.find(class_="language-json")
        j = json.loads(js.text)
        newdict = {}
        for key,value in j['Npi'].items():
            if key == 'Taxonomys':
                for i in j['Npi']['Taxonomys']['Taxonomy']:
                    newdict[i] = j['Npi']['Taxonomys']['Taxonomy'][i]
            else:
                newdict[key] = j['Npi'][key]
        the_list.append(newdict) 
    except:
        failed_npi.append(npi)
        print (npi)
```

```python
keys = the_list[0].keys()
with open(r'C:\Users\vkashya5\Desktop\NPI.csv', 'w+') as output_file:
    dict_writer = csv.DictWriter(output_file, fieldnames=keys)
    dict_writer.writeheader()
    dict_writer.writerows(the_list)
```
